```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], kde=True, bins=25)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
#sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')  # kind='reg' - regerssion
```

```python
sns.pairplot(data=tips, hue='day') # hue allows you to add a categorical column and plot depicts them in the 
                                   # diagram e.g. day shows the 7 days in the week. 
```

```python
sns.pairplot(data=tips, hue='day', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill']) # more dense the lines means more data. 
```

```python
sns.kdeplot(tips['total_bill'])
```
