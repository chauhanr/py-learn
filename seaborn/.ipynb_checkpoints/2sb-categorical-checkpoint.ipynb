```python
import seaborn as sns 
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# The categorical plot will concern with columns that represent categories e.g. Sex, Smoker, Day etc. 
# the most popular of the categorical plot is bar 
sns.barplot(x='sex', y='total_bill', data=tips)

# x should have the category column and the y as the numerical column. WE can interchange x and y but we will
# get a horizontal bar in that case. 
# graph below shows mean of the total_bills for males vs females. which suggests on average males spend more 
# than females. 
```

```python
import numpy as np 
```

```python
# by default the barplot plots the mean of the numberical column for all the categories 
# if you have to use a different estimator func we can use the estimator parameter in the barplot 
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std) # here we use the standard deviation as the est
```

```python
# count plot is the next categorical plot except that y column is already decided and is sum 
sns.countplot(x='day', data=tips)
```

## Box Plots - distribution of categorical data 

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

## Violin Plot 

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

## Swarm plot 

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```
