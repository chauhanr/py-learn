Here we have the images in one folder and the outline of the image in another. 

```python
import os 

# extarct the paths so that we know which input image maps to the target annotation.
input_dir="images/"
target_dir="annotations/trimaps/"

input_img_paths = sorted(
    [os.path.join(input_dir, fname) 
     for fname in os.listdir(input_dir)
     if fname.endswith(".jpg")])
target_paths = sorted(
    [os.path.join(target_dir, fname)
     for fname in os.listdir(target_dir)
     if fname.endswith(".png") and not fname.startswith(".")])
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.utils import load_img, img_to_array
```

```python
# Display input image #7
display_num = 10
img = load_img(input_img_paths[display_num])
plt.imshow(img)
plt.axis("off")
```

```python
def display_target(target_array):
    normalized_array = (target_array.astype("uint8") - 1) * 127         
    plt.axis("off")
    plt.imshow(normalized_array[:, :, 0])

tar_img = img_to_array(load_img(target_paths[display_num], color_mode="grayscale"))   
display_target(tar_img)
```
