```python
from matplotlib import pyplot as plt
import numpy as np 
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=1,linestyle='-', marker='o', 
        markersize=10, markerfacecolor='red', markeredgecolor='black', markeredgewidth=1)  
# also give color= grene
# linewidth (lw)
# alpha - transparence
# linestyle (ls) --, -., etc
# marker (+, x, o)
```

```python
# limits 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=1,linestyle='--')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
