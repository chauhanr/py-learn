```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for curr_ax in axes: 
    curr_ax.plot(x,y)
```

```python
# same thing with index
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title("Plot 1")
axes[1].plot(y,x)
axes[1].set_title("Plot 2")
```

# Figure Size and DPI

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,4), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='squared')
ax.plot(x,x**3, label='cubed')

ax.legend(loc=2)
```
