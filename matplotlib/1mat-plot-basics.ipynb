```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
## Object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width and height of the axes (l,b,w,h)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Parabola")

axes.plot(x,y)
```

```python
# putting more plots on a canvas 
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_xlabel("X1")
axes1.set_ylabel("Y1")
axes2 = fig.add_axes([0.2,0.4,0.4,0.4])

axes1.plot(x,y)
axes2.plot(y,x)
```
